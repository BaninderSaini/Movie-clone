```python
main_data.dropna(inplace=True)
main_data.info()
```

```python
#main_data = main_data[['movieId','tmdbId','title','genres','rating','tag']]
main_data = main_data[['movieId','tmdbId','title','genres','tag']]

main_data["Year"]=main_data["title"].str[-5:-1:1]
main_data
```

```python
main_data["title"]=main_data["title"].str[:-7]
main_data
```

```python
main_data["genres"]=main_data["genres"].str.split('|')
```

```python
main_data
```

```python
main_data['tag'] = main_data['tag'].apply(lambda x:x.split())
```

```python
main_data
```

```python
main_data["tag"]=main_data["tag"]+main_data["genres"]
```

```python
main_data
```

```python
main_data['tag'] = main_data['tag'].apply(lambda x: " ".join(x))
```

```python
main_data.info()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c = CountVectorizer(max_features=1000,stop_words='english')
```

```python
v = c.fit_transform(main_data['tag']).toarray()
v
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
s = cosine_similarity(v)
```

```python
def func(movie):
    ind = main_data[main_data['title'] == movie].index[0]
    d = sorted(list(enumerate(s[ind])),reverse=True,key = lambda x: x[1])
    for i in d[0:6]:
        print(main_data.iloc[i[0]].title)
```
